kolumna BODYTYPE określająca typ pojazdu - do analizy bierzemy wyłącznie kategorię 0 oraz 5

```python
new_df["BODYTYPE"].value_counts()
```

```python
new_df_odfiltrowana=new_df[(new_df["BODYTYPE"]==0) | (new_df["BODYTYPE"]==5)]
```

```python
new_df_odfiltrowana
```

```python
pasy_vs_obrazenia=new_df_odfiltrowana.groupby(["MANUSE","INJSEV"])["MANUSE","INJSEV"].count()
pasy_vs_obrazenia
```

```python
pasy_vs_obrazenia2=new_df_odfiltrowana.groupby(["MANUSE","INJSEV"])["MANUSE"].count()
pasy_vs_obrazenia2
```

```python
pasy_vs_obrazenia2.sum()
```

```python
prawodpodobienstwo_obrazen_vs_pasy=pasy_vs_obrazenia2/pasy_vs_obrazenia2.sum()
```

```python
prawodpodobienstwo_obrazen_vs_pasy
```

MANUSE: 1 - brak pasów, 2 - były pasy

INJSEV
0 - brak obrażeń, 
1 - możliwe obrażenia,
2 - nieobezwładniające obrażenia,
3 - obezwładniające obrażenia,
4 - śmierć,
5 - wielkość obrażeń nie jest znana, 
6 - śmierć przed wypadkiem


```python
new_df_odfiltrowana[["MANUSE","INJSEV"]].corr(method='spearman')
```

```python
df_task3=pd.DataFrame(prawodpodobienstwo_obrazen_vs_pasy)
```

```python
df_task3
```

```python
df_task3["RESULT"]=df_task3["MANUSE"]
```

```python
df_task3.drop(columns=["MANUSE"], inplace = True)
```

```python
df_task3
```

```python
pd.to_numeric(df_task3['RESULT'])
```
